# Data Reading and Writing and Creating Delta Talbes

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

#### trip data && trip zone

```python
silver = "abfss://silver@nyctaxicesa.dfs.core.windows.net"
gold = "abfss://gold@nyctaxicesa.dfs.core.windows.net"
```

```python
df_trip_zone = spark.read.format("parquet")\
                    .option("header", "true")\
                    .option("inferSchema", "true")\
                    .load(f"{silver}/trip-zone")
```

```python
df_trip_zone.display()
```

```python

gold = "abfss://gold@nyctaxicesa.dfs.core.windows.net"

# Write the DataFrame as a Delta table in the gold database
df_trip_zone.write.format('delta') \
    .mode('append') \
    .option("path", f"{gold}/trip-zone") \
    .saveAsTable("gold.`trip-zone`")

```

```python
df_trip_zone.display()
```

```python
df_trip_data = spark.read.format("parquet")\
                    .option("header", "true")\
                    .option("inferSchema", "true")\
                    .load(f"{silver}/trip-data")
```

```python
%sql 
SELECT * FROM gold.`trip-zone`
WHERE zoen1 = 'Astoria'
```

```python
df_trip_data.display()
```

```python

gold = "abfss://gold@nyctaxicesa.dfs.core.windows.net"
# Write the DataFrame as a Delta table in the gold database
df_trip_data.write.format('delta') \
    .mode('append') \
    .saveAsTable("trip_data")
```

```python
%sql  
SELECT * FROM gold.`trip-data`
```


## Trip Data 2023

```python
df_trip_data_2023 = spark.read.format("parquet")\
                    .option("header", "true")\
                    .option("inferSchema", "true")\
                    .load(f"{silver}/trip-data-2023")
```

```python
display(df_trip_data_2023)
```

```python

gold = "abfss://gold@nyctaxicesa.dfs.core.windows.net"

df_trip_data_2023.write.format('delta') \
    .mode('overwrite') \
    .saveAsTable("gold.trip_data_2023")

```
